```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
# the four different states of the And gate
training_data = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]).astype("float32")

# the four expected results in the same order
target_data = np.array([
    [0],
    [1],
    [1],
    [1]
]).astype("float32")
```

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
model.fit(training_data, target_data, nb_epoch=500, verbose=1)
```

```python
hasil = model.predict(training_data)
print('Hasil Mikir : ')
print(hasil)
print()

bulat = hasil.round()
print('Pembulatan  : ')
print(bulat)
```
